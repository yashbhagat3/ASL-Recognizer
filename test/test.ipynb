```python
img = cv2.imread('testw.jpg')
blur = cv2.GaussianBlur(img,(5,5),0)
gray = cv2.cvtColor(blur,cv2.COLOR_BGR2GRAY)


for i in range(200):
    for j in range(200):
        if(gray[i,j])>128:
            gray[i,j]=0
        else:
            gray[i,j]=255
            
plt.imshow(gray,cmap = 'gray')
```

```python
hand_cascade = cv2.CascadeClassifier('hand.xml')
hands = hand_cascade.detectMultiScale(img)

for (x,y,w,h) in hands:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)

plt.imshow(img,cmap='gray')
```

```python
equ = cv2.equalizeHist(gray)
plt.imshow(gray,cmap='gray')
plt.show()
plt.imshow(equ,cmap='gray')
plt.show()
# cv2.imshow('Frame', equ)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
img1=cv2.Canny(gray,200,100)
plt.imshow(img1,cmap='gray')
plt.show()

img2=cv2.Canny(equ,200,100)

plt.imshow(img2,cmap='gray')
```

```python
img = cv2.imread('test3.jpg')
mask = np.zeros(img.shape[:2],np.uint8)

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (50,50,100,100)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,15,cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
 
plt.imshow(img,cmap='gray')
```

```python
img = cv2.imread('test3.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blured = cv2.GaussianBlur(gray, (5, 5), 0)
blured = cv2.erode(blured, None, iterations=1)
blured = cv2.dilate(blured, None, iterations=0)
lower = cv2.getTrackbarPos('lower_threshold', 'edged')
upper = cv2.getTrackbarPos('upper_threshold', 'edged')
# edged = cv2.Canny(blured,lower,upper)
edged = cv2.Canny(blured,50,100)

plt.imshow(blured,cmap='gray')
plt.show()
plt.imshow(edged,cmap='gray')
plt.show()
```
